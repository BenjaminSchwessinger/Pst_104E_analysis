```python
import requests, sys
 
server = "http://rest.ensemblgenomes.org"
ext = "/info/genomes/division/EnsemblFungi?"
 
r = requests.get(server+ext, headers={ "Content-Type" : "application/json"})
 
if not r.ok:
  r.raise_for_status()
  sys.exit()
 
decoded = r.json()
print(repr(decoded))
```

```python
import re
ftp_short = 'ftp://ftp.ensemblgenomes.org/pub/fungi/release-36/fasta/%s/cdna/%s.%s.cdna.all.fa.gz'
ftp_long = 'ftp://ftp.ensemblgenomes.org/pub/fungi/release-36/fasta/%s/%s/cdna/%s.%s.cdna.all.fa.gz'
new_database = []
for x in decoded:
    db = re.findall(r'(.*)_core.*', x['dbname'])[0]
    species = x['species']
    assembly = x['assembly_name']
    if any(x for x in [' ', '.', ',', '/'] if x in assembly):
        assembly = assembly.replace(' ', '_').replace(',','_').replace('.','_').replace('__','_')
        assembly = assembly.replace('/','_')
    if assembly == 'EF 1':
        assembly = assembly = 'EF1'
    if 'JCVI' in assembly:
        assembly = assembly.replace('JCVI','CADRE')
    if species == db:
        url = ftp_short % (db, species.capitalize(), assembly)
        new_database.append(url)
    elif species != db:
        url = ftp_long % (db, species ,species.capitalize(), assembly)
        new_database.append(url)
```

```python
import os
os.chdir('/home/benjamin/databases/EnsembelFungi/0817_download')
worked = []
not_worked = []
count = 0
to_download = len(new_database)
stderr_error_dict = {}
for wget in new_database:
    stderr = !wget {wget}
    count += 1
    stderr_error_dict[wget] = stderr
    print('Downloaded? %i out of %i databases %i to go' %(count, to_download, to_download-count ))
    #print(stderr)
    #implant = 0
    if any(x for x in stderr if 'saved' in  x):
        worked.append(wget)
    else:
        not_worked.append(wget)
        print("Have not worked so far: ", not_worked)
```

```python
import re
import time
import time
import ftplib
ftp = ftplib.FTP("ftp.ensemblgenomes.org")
ftp.login("anonymous", "")
total_miss = []
count = 0
for miss in not_worked[29:]:
    files = []
    if '/4D' in miss:
        miss = miss.replace('/4D','_4D')
        
    path = re.findall(r'ensemblgenomes.org/(.*)/', miss)[0] + '/'
    try:
        files = ftp.nlst(path)
    except ftplib.error_perm :
            print ("No files in this directory")
    cdna_path = [x for x in files if x.endswith('cdna.all.fa.gz')][0]
    time.sleep(2)
    print(cdna_path)
    stderr = !wget ftp://ftp.ensemblgenomes.org/{cdna_path}
    if any(x for x in stderr if 'saved' in  x):
        worked.append(miss)
        count += 1
        print('Out of %is misses we downloaded %i' % (len(total_miss), count))
    else:
        total_miss.append(miss)
        print("Have not worked so far: ", total_miss)
ftp.close()
```
