```python
import pandas as pd
import numpy as np
```

```python
gff_p = pd.read_csv('../122016_assembly/Pst_E104_v1_p_ctg.anno.gff3', sep='\t',skiprows=1, header=None)
```

```python
p_filter_array= np.loadtxt('../TE_polishing/Pst_E104_v1_p_ctg.RepaseTPSI_filter.txt',dtype=str)
```

```python
p_filter_array[0][12:-1]
```

```python
#this needs fixing to incooporate exons as well
p_string = p_filter_array[0][12:-1]+';|'+p_filter_array[0][12:-1]+".exon"
for x in p_filter_array:
    p_string = p_string +'|' + x[12:-1]+';|'+x[12:-1]+'.exon'
```

```python
p_bool_filter_array = gff_p[8].str.contains(p_string)
```

```python
len(p_filter_array) == p_bool_filter_array.count()
```

```python
p_bool_filter_array.value_counts()
```

```python
gff_filtered_out = gff_p[p_bool_filter_array]
```

```python
gff_filtered_in = gff_p[~p_bool_filter_array]
```

```python
gff_filtered_out.to_csv('../TE_polishing/Pst_E104_v1_p_ctg.anno.RepaseTPSI.gff3', sep='\t', header=None, index=None)
```

```python
gff_filtered_in.to_csv('../TE_polishing/Pst_E104_v1_p_ctg.anno.RepaseTPSI_filtered.gff3', sep='\t', header=None, index=None)
```

```python
from Bio import SeqIO
p_filtered_protein_id_list = pd.read_csv('../TE_polishing/Pst_E104_v1_p_ctg.RepaseTPSI_filter.txt', sep='\t', header=None)[0].tolist()
p_filtered_protein_list =[]
for seq in SeqIO.parse(open('../122016_assembly/Pst_E104_v1_p_ctg.anno.protein.fa'), 'fasta'):
    if seq.id not in p_filtered_protein_id_list:
        seq.description = ''
        seq.name =''
        p_filtered_protein_list.append(seq)
out_handle = open('../122016_assembly/Pst_E104_v1_p_ctg.anno.RepaseTPSI_filtered.protein.fa', 'w')
SeqIO.write(p_filtered_protein_list, out_handle, 'fasta')
```
